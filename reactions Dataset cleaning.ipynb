```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/Users/macuser/Documents/accenture INt/Reactions.csv')
df
```

```python
df.head()
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
```

```python
df['month'] = df['Datetime'].dt.strftime('%Y-%m')
```

```python
monthly_post_counts = df['month'].value_counts()
```

```python
most_posts_month = monthly_post_counts.idxmax()
most_posts_count = monthly_post_counts.max()
```

```python
monthly_post_counts.plot(kind='bar')
plt.xlabel("Month")
plt.ylabel("Number of Posts")
plt.title("Number of Posts per Month")
plt.tight_layout()
plt.show()

print("Month with the most posts:", most_posts_month)
print("Number of posts in the most active month:", most_posts_count)



```
